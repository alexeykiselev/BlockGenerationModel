```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

% matplotlib inline
plt.style.use('thinner.mplstyle')

C = 1000000
x = np.linspace(1, C, num=100)
y = x
a = y / C

b = np.fabs(np.log2(a))

v = np.random.randint(0, C, 10000)

d1 = stats.gaussian_kde(v)
x1 = np.linspace(0, C, 100)
plt.plot(x1, d1(x1))
plt.hist(v, bins=100, normed=True)
plt.show()

# d = b * (C / np.log2(C)) 
# dd = b * C
# ddd = b * C / 2
# plt.plot(x, d)
# plt.show()

cv = np.fabs(np.log2(v)) * (C / np.log2(C))
d2 = stats.gaussian_kde(cv)
x2 = np.linspace(0, C, 100)
plt.plot(x2, d2(x2))
plt.hist(cv, bins=100, normed=True)
plt.show()

cv2 = 2 * (C - ((2 / C) * v * v - (2 / C) * C * v + C))
d3 = stats.gaussian_kde(cv2)
x3 = np.linspace(0, C, 100)
plt.plot(x3, d3(x3))
plt.hist(cv2, bins=100, normed=True)
plt.show()

```
